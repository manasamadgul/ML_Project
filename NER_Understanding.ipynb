**Named Entity Recognition using Nltk calculating precion and recall**

```python
from transformers import pipeline

# Load the NER pipeline
ner_classifier = pipeline("ner",grouped_entities=True)

# Example sentence
sentence = "Manasa is in Orlando, manasamadgul@gmail.com John and Mary went to the park. Alice and Bob went to the store. John Smith and Mary Johnson are getting married next week.David and Emily are best friends since childhood.Michael Jackson was a legendary pop singer.Sarah and Matthew are siblings."

# Use the NER pipeline to extract named entities
ner_results = ner_classifier(sentence)
print(ner_results)
# Extract only person names
extracted_names = [entity['word'] for entity in ner_results if entity['entity_group'] == 'PER']

print("Extracted Names:", extracted_names)

# Ground truth dataset (manually annotated names for the sentence)
ground_truth_names = ["John", "Mary", "Alice", "Bob","John Smith", "Mary Johnson", "David", "Emily", "Michael Jackson", "Sarah", "Matthew"]
#ground_truth_names = extracted_names

# Calculate precision and recall
true_positives = len(set(extracted_names) & set(ground_truth_names))
false_positives = len(set(extracted_names) - set(ground_truth_names))
false_negatives = len(set(ground_truth_names) - set(extracted_names))

precision = true_positives / (true_positives + false_positives) if true_positives + false_positives > 0 else 0
recall = true_positives / (true_positives + false_negatives) if true_positives + false_negatives > 0 else 0

print("Precision:", precision)
print("Recall:", recall)
```

**Named Entity Recognition using Spacy calculating precion and recall**

```python
import spacy

# Load the English language model
nlp = spacy.load("en_core_web_sm")

# Function to extract names from a sentence using spaCy
def extract_names(sentence):
    doc = nlp(sentence)
    persons = [ent.text for ent in doc.ents if ent.label_ == "PERSON"]
    return persons

# Function to calculate precision and recall
def calculate_precision_recall(extracted_names, ground_truth_names):
    true_positives = len(set(extracted_names) & set(ground_truth_names))
    false_positives = len(set(extracted_names) - set(ground_truth_names))
    false_negatives = len(set(ground_truth_names) - set(extracted_names))

    precision = true_positives / (true_positives + false_positives) if true_positives + false_positives > 0 else 0
    recall = true_positives / (true_positives + false_negatives) if true_positives + false_negatives > 0 else 0

    return precision, recall

# Example sentence
sentence = "Manasa is in Orlando. Ahana is at school , John and Mary went to the park. Alice and Bob went to the store. John Smith and Mary Johnson are getting married next week.David and Emily are best friends since childhood.Michael Jackson was a legendary pop singer.Sarah and Matthew are siblings."

# Extract names from the sentence using spaCy
extracted_names = extract_names(sentence)

# Ground truth dataset (manually annotated names for the sentence)
ground_truth_names = ["John", "Mary", "Alice", "Bob","John Smith", "Mary Johnson", "David", "Emily", "Michael Jackson", "Sarah", "Matthew"]

# Calculate precision and recall
precision, recall = calculate_precision_recall(extracted_names, ground_truth_names)

print("Extracted Names:", extracted_names)
print("Precision:", precision)
print("Recall:", recall)
```
